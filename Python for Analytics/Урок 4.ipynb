## Решение

Чтобы изучить характеристики домов, влияющие на их цену, можно выполнить следующие шаги:

- Создать гистограмму, показывающую распределение цен на дома. Эта гистограмма показывает распределение цен на дома в наборе данных. Можно заметить, что большинство домов стоят от 200 000 до 800 000 долларов.
- Создать диаграмму рассеяния, показывающую взаимосвязь цены на дом и площади. Эта диаграмма рассеяния показывает, что существует положительная корреляция между ценой на дом и площадью жилой площади, что означает, что дома с большей площадью жилой площади обычно стоят дороже.
- Создать гистограмму, показывающую распределение домов по количеству спален. Эта диаграмма рассеяния показывает, что существует положительная корреляция между ценой на дом и количеством спален, что означает, что дома с большим количеством спален обычно стоят дороже.


```python
import seaborn as sns
sns.scatterplot(x='price', y='sqft_living', data=df)
plt.xlabel('Цена')
plt.ylabel('Площадь жилой площади')
plt.show()
```

## Вывод

Распределение цен на дома в наборе данных скошено влево, с большинством домов стоящих от 200 000 до 800 000 долларов. Существует положительная корреляция между ценой на дом и площадью жилой площади, что означает, что дома с большей площадью жилой площади обычно стоят дороже.

```python
sns.scatterplot(x='price', y='bedrooms', data=df)
plt.xlabel('Цена')
plt.ylabel('Количество спален')
plt.show()
```

## Вывод

Большинство домов имеет 3 или 4 спальни. Существует положительная корреляция между ценой на дом и количеством спален, что означает, что дома с большим количеством спален обычно стоят дороже.

```python
sns.jointplot(x='grade', y='price', data=df)
plt.show()
```

## Вывод

Исследование показало, что стоимость недвижимости растет с ростом оценочного класса (grade). Однако эта зависимость не является строго линейной, и наблюдается разброс цен внутри каждого класса. Возможные причины этого - район, наличие вида, размер и т.д. В целом тенденция к росту цен с ростом оценочного класса указывает на то, что недвижимость высокого класса ценится больше.


```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5)

plt.xlabel('Цена')
plt.ylabel('Вид на воду')
plt.title('Распределение цены в зависимости от вида на воду')
```

## Вывод

На представленном графике видно, что стоимость недвижимости с видом на воду существенно выше, чем у недвижимости без вида на набережную. Такая недвижимость считается более желанной из-за своего расположения и вида на воду, что делает ее более ценной.

```python
plt.figure(figsize=(16, 8))

sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5)

plt.xlabel('Цена')
plt.ylabel('Вид')
plt.title('Распределение цены в зависимости от вида');
```

## Вывод
На представленном графике видно, что стоимость недвижимости зависит от уровня вида. Недвижимость с высоким уровнем вида (категория "4") ценится существенно выше, чем недвижимость с низким уровнем вида (категория "0").
